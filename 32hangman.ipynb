# 32. Hangman

From [PracticePython.org](https://www.practicepython.org/exercise/2017/01/10/32-hangman.html):

> In this exercise, we will finish building Hangman. In the game of Hangman, the player only has 6 incorrect guesses (head, body, 2 legs, and 2 arms) before they lose the game.
>
> [Read full description](https://www.practicepython.org/exercise/2017/01/10/32-hangman.html)

Gratitude to [Chris Horton](https://gist.github.com/chrishorton) for the ascii art!

```python
import random as rd

with open("assets/sowpods.txt") as file:
    words = file.read().split()

def get_word():
    return rd.choice(words)

def print_board():
    print(gallows[wrong])
    print(" ".join(board))
    print()

playing = True

# ascii hangman art from https://gist.github.com/chrishorton/8510732aa9a80a03c829b09f12e20d9c 
gallows = ['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']
    
while playing:
    word = get_word()
    guesses = set()
    wrong = 0
    board = ["_" for char in list(word)]

    print("Welcome to Hangman!")
    print_board()

    while wrong < 6:
        correct = False
        guess = input("Guess your letter: ").upper()
        if guess == "":
            playing = False
            break
        if guess in guesses:
            print("You already guessed that!")
            continue
        guesses.add(guess)
        for i in range(len(word)):
            if guess == word[i]:
                correct = True
                board[i] = guess
        if not correct:
            wrong += 1
        print_board()
        if "_" not in board:
            print("You win!")
            break
    else:
        print("You lose!\nThe word was", word)
    if playing:
        again = input("Play again? (Y/N) ")
        if again.lower() != "y":
            playing = False
```
